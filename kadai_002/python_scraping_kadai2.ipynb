```python
import requests
import pprint

from getpass import getpass
api_key = getpass("APIキーを入力してください\n")

serch_station = input("駅名を入力してください\n")
```

```python
# 駅の情報を取得
station_url = "https://maps.googleapis.com/maps/api/geocode/json"

response_station = requests.get(
    station_url,
    params={
        "address": serch_station,
        "key": api_key,
    }
    )

lat_station = response_station.json()["results"][0]["geometry"]["location"]["lat"]
lng_station = response_station.json()["results"][0]["geometry"]["location"]["lng"]

# 駅
json_station_data = response_station.json()
pprint.pprint(json_station_data)

print(f"緯度は{lat_station}です")
print(f"経度は{lng_station}です")
```

```python
# レストランの情報を取得
url_restaurant = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"


response_restaurant = requests.get(
    url_restaurant,
    params={
        "location": f"{lat_station},{lng_station}", # 緯度と経度を結合して 'location' パラメータに設定
        "language": "ja", # 言語
        "radius": 500, # 半径
        "type": "restaurant",
        "key": api_key
        # 'range' パラメータはGoogle Places API Nearby Searchでは無効なので削除します
    }
)

json_restaurant_data = response_restaurant.json()
pprint.pprint(json_restaurant_data)
```

```python
# レストラン名
# 評価
# 住所
for shop in json_restaurant_data["results"]:
  name = shop.get("name")
  rating = shop.get("rating")
  address = shop.get("vicinity")
  print(f"店名: {name}")
  print(f"評価: {rating}")
  print(f"住所: {address}")
  print("------")
```
