```python
import requests
from bs4 import BeautifulSoup
# 取得するURL
url = "https://news.yahoo.co.jp/articles/922f1ffe2957dd6751df976723f09e4eaaa93aed"
# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)
# HTMLを解析する
soup = BeautifulSoup(response.text, "html.parser")
# 記事の本文を抽出する
p_element_1 = soup.select_one('#uamods > div.article_body.highLightSearchTarget > div:nth-child(1) > p')
# a タグだけを削除
if p_element_1: # p_element_1 の中からaタグを削除
    for a in p_element_1.select('a'):
        a.decompose()
    p_element_3 = p_element_1.get_text(strip=True)
else:
    p_element_3 = "" # If p_element_1 is not found, set p_element_3 to an empty string

p_element_2 = soup.select_one('#uamods > div:nth-of-type(1) > div:nth-of-type(2) > p')
# 記事を表示する
print(p_element_3)

if p_element_2: # Check if p_element_2 was found
    print(p_element_2.text)
else:
    print("p_element_2 not found.")
```
