```python
# Seleniumで必要なライブラリをインポート（何を使うかによってインポートする種類も異なる）。

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from getpass import getpass
```

```python
# ヘッドレスモードで起動するためのオプションを設定
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# ChromeDriverManagerを使用してChromeDriverを自動的に管理
service = Service(ChromeDriverManager().install())

# Google Chrome用のブラウザドライバーインスタンスを作成
chrome_driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
# TERAKOYAのトップページにアクセス
chrome_driver.get('https://terakoya.sejuku.net')

# 最大30秒間、ログインボタンが表示されるのを待つ
wait = WebDriverWait(chrome_driver, 30)
login_button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,
'#root > header > div > div > div.sc-cdmAjP.cJvsei > div.sc-jytpVa.sc-pYNGo.ljsdtk.hsShUO')))

# ログインボタンをクリック
login_button.click()
```

```python
# メールアドレスとパスワードを入力し、変数にそれぞれ格納しておく
email_address = input('メールアドレスを入力してください：')
password = getpass('パスワードを入力してください：')
```

```python
# メールアドレスとパスワードの入力欄を探す
email_input = chrome_driver.find_element(By.NAME, 'email')
password_input = chrome_driver.find_element(By.NAME,'password')
```

```python
# 入力したメールアドレスとパスワードを入力欄に入れる
email_input.send_keys(email_address)
password_input.send_keys(password)
```

```python
# ログインボタンをクリックしてログイン

form_login_button = wait.until(
    EC.visibility_of_element_located(
        (By.CSS_SELECTOR,
         '#root > div.sc-kNOymR.TvzZn > div.sc-lgpSej.gsAReM > div.sc-dntSTA.dQOsDP > button')
    )
)

form_login_button.click()
```

```python
# マイページにアクセスしたら、右上のアイコンボタンを探してクリック
icon_butttom = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,
                                             '#root > div.sc-kybZmf.cQgDNh > div > div > header > div.sc-flkahu.iJMNgb > div.sc-dErsJf.jacRqp > div')))
icon_butttom.click()

# アカウント設定ボタンを探してクリック
account_buttom = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,
                                            '#root > div.sc-kybZmf.cQgDNh > div > div > header > div.sc-dua-dxq.dsONQN > ul > a:nth-child(1) > li')))
account_buttom.click()
```

```python
# 最大30秒間、編集ボタンが表示されるのを待つ
wait = WebDriverWait(chrome_driver, 30)
edit_button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,
'#root > div.sc-hmBdrb.jqAYfT > div > div > main > div > div.sc-bVsQZF.eQwtkq > button')))
# 編集ボタンをクリック
edit_button.click()
```

```python
comment = input('自己紹介文を記入してください：')
```

```python
# 自己紹介文の入力欄を探してクリック
comment_input = chrome_driver.find_element(By.CSS_SELECTOR,
                                           '#root > div.sc-hmBdrb.jqAYfT > div > div > main > div > div.sc-bVsQZF.eQwtkq > div:nth-child(10) > div.sc-jsbVYh.ljsGeL > textarea')
comment_input.click()
```

```python
# 自己紹介文の文言を入れる

comment_input.send_keys(comment)
```

```python
# 更新ボタンを探してクリック
update_button = chrome_driver.find_element(By.CSS_SELECTOR,
                                           '#root > div.sc-hmBdrb.jqAYfT > div > div > main > div > div.sc-bVsQZF.eQwtkq > button.sc-dTvVRJ.bHRgJQ.sc-lfzvHA.kmNVuB')

update_button.click()
```

```python
# ブラウザを閉じる
chrome_driver.quit()
```
